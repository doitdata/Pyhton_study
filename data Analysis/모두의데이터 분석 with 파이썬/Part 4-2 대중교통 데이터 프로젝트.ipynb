```python
import csv 
f = open('d:/Python_Project/data_sample/Unit12/subwaytime.csv')
data = csv.reader(f)
next(data) # 1행 헤더값 제외하기
next(data) #  2행 헤더값 제외하기

for row in data:
    #4번 인덱스 부터 전체 값 정수형으로 데이터 변환
    # map()함수는 일괄적으로 데이터에 특정함수를 적용할 수 있음
    # 첫번쨰 인자에 일괄 적용할 함수 이름을 입력하고,
    #두번째 인자에 그 함수를 적용할 데이터를 입력
    row[4:] = map(int, row[4:]) 
    print(row)
```

```python
# 밤 11시ㅔ 사람들이 가잔 많이 타고 내리는 역 찾기
import csv 
f = open('d:/Python_Project/data_sample/Unit12/subwaytime.csv')
data = csv.reader(f)
next(data) # 헤더값 제외하기
next(data) 
mx=0
mx_station=''
t = int(input('몇시의 승차인원이 가장 많은 역이 궁금하세요? : '))
for row in data:
    row[4:] = map(int, row[4:])
    a=row[2*t-4]
    if a > mx:
        mx = a
        mx_station = row[3] + '(' + row[1] +')'
print(mx_station, mx)
```

```python
#시간대별로 사람들이 가장많이 타고 내리는 역
import csv 
f = open('d:/Python_Project/data_sample/Unit12/subwaytime.csv')
data = csv.reader(f)
next(data) # 헤더값 제외하기
next(data) 
mx=[0]*24 #시간대별 최대 승차인원 저장 리스트 초기화
mx_station=['']*24 #시간대별 최대 승차인원  역이름 저장 리스트 초기화 
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24): # 전체 24시간이니까 0부터 23인덱스
        a=row[j*2+4] #패턴입력
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]
            
print(mx_station)
print(mx)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station,rotation=90) #역이름을 90도로 회전
plt.show()
```

```python
#시간대별로 하차인원이 가장 많은 역을 찾는 코드
import csv 
f = open('d:/Python_Project/data_sample/Unit12/subwaytime.csv')
data = csv.reader(f)
next(data) # 헤더값 제외하기
next(data) 
mx=[0]*24
mx_station=['']*24
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        b=row[5+j*2] # j값과 인덱스 번호 값의 관계식 사용
        if b > mx[j]:
            mx[j] = b
            mx_station[j] = row[3]+'('+str(j+4)+')' 
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv 
f = open('d:/Python_Project/data_sample/Unit12/subwaytime.csv')
data = csv.reader(f)
next(data) # 1행 헤더값 제외하기
next(data) # 2행 헤더값 제외하기
s_in=[0]*24
s_out=[0]*24
for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4+i*2] # 시간대별 승차인원 누적
        s_out[i] += row[5+i*2] #시간대별 하차인원 누적
        
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
#여기서 1e7 => 1 곱하기 10의 승으로 y축 단위는 천망명임을 알 수 있다. 
```
